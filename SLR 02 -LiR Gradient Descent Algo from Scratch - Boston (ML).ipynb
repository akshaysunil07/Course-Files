```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_boston
```

```python
import statsmodels.api as sm
```

```python
boston=load_boston()
```

```python
df=pd.DataFrame(boston.data,columns=boston.feature_names)
df['Price']=boston.target
df.head()
```

```python
y=df['Price']
x=df['LSTAT']
```

```python
epochs= 10000
n=len(x)
m=0.5
c=10
L=0.0001
for i in range(epochs):
    y_pred=m*x+c
    D_m=(-2/n)*sum(x*(y-y_pred))
    D_c=(-2/n)*sum(y-y_pred)
    m=m-L*D_m
    c=c-L*D_c
    
    if i%1000==0:
        plt.scatter(x,y)
        plt.plot(x,y_pred,'r')
        plt.show()

print('m =',m,' c =',c)
# Epochs means no.of times the entire data was read.
```

```python
# We didn't observe much change, so we change the learning rate.
epochs= 10000
n=len(x)
m=0.5
c=10
L=0.001
for i in range(epochs):
    y_pred=m*x+c
    D_m=(-2/n)*sum(x*(y-y_pred))
    D_c=(-2/n)*sum(y-y_pred)
    m=m-L*D_m
    c=c-L*D_c
    
    if i%1000==0:
        plt.scatter(x,y)
        plt.plot(x,y_pred,'r')
        plt.show()

print('m =',m,' c =',c)
```
