# Logistic Regression Model

```python
y=df['TenYearCHD']
X=df.drop('TenYearCHD',axis=1)
```

```python
import statsmodels.api as sm
Xc=sm.add_constant(X)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
```

```python
pd.DataFrame([vif(Xc.values,i)for i in range(Xc.shape[1])],index=Xc.columns,columns=['VIF'])
```

```python
model = sm.Logit(y,Xc).fit()
model.summary()
```

```python
model.pvalues.max()
```

```python
pdf=pd.DataFrame(model.pvalues,columns=['pval'])
```

```python
cols=list(pdf[pdf['pval']<=0.05].index)
```

```python
Xc=Xc[cols]
model = sm.Logit(y,Xc).fit()
model.summary()
```

```python
param=pd.DataFrame(model.params,columns=['Coeff'])
param['ExpCoeff']=np.exp(model.params)
```

```python
param
```

```python
pd.DataFrame((param['ExpCoeff']-1)*100)
```

```python
Xc2=Xc[cols]
Xc2['CHD']=y
```

```python
Xc2.to_excel('Chd_Excel.xlsx')
```

```python
prob = model.predict(Xc[cols])
prob = pd.DataFrame(prob,columns=['Probability'])
prob['CHD']=y
prob['y_est']=prob['Probability'].apply(lambda x: 0 if x<0.5 else 1)
prob.head()
```

```python
prob[prob['y_est']==1]
```

```python
print('Number of Obs with CHD and corrected predicted CHD for them:')
print(prob[(prob['y_est']==1) & (prob['CHD']==1)].shape[0])
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(prob['CHD'],prob['y_est'])
```

```python
print('Accuracy of the model:')
print(accuracy_score(prob['CHD'],prob['y_est']))
```

```python
sns.distplot(prob['Probability'][prob["CHD"]==0],color='g',label='P-Distb of 0')
sns.distplot(prob['Probability'][prob["CHD"]==1],color='r',label='P-Distb of 1')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(prob['CHD'],prob['y_est']))
```

## Logistic Regression - Machine Learning

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(solver='liblinear')
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
lr.fit(X_train,y_train)
y_train_pred=lr.predict(X_train)
y_train_prob=lr.predict_proba(X_train)[:,1]

print('Confusion Matrix Train\n',confusion_matrix(y_train,y_train_pred))
print('\nAccuracy Train:',accuracy_score(y_train,y_train_pred))


y_test_pred=lr.predict(X_test)
y_test_prob=lr.predict_proba(X_test)[:,1]

print('\n\nConfusion Matrix Test\n',confusion_matrix(y_test,y_test_pred))
print('\nAccuracy Test:',accuracy_score(y_test,y_test_pred))
```
